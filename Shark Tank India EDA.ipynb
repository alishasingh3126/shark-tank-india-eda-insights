```r
#Hypothesis testing (Example: t-test)
t_test= t.test(df$ask_equity, df$deal_equity)
print(t_test)
```

To determine whether to accept or reject the null hyporhesis:

* If the p-value is less than the significance level (commonly chosen as 0.05), we reject the null hypithesis.
* If the p-value is greater than the significance level, we fail to reject the null hypothesis.

In this case, the p-value (0.003343) is less than 0.05, so we reject the null hypothesis. Therefore, we conclude that there is a statistically significant difference between the means of ask_equity and deal_equity
